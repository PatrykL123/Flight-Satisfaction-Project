# <h1>DATA MAPPING</h1>

```python
satisfaction_mapping = {'Neutral or Dissatisfied':0,'Satisfied':1}
```

```python
raw_data_two['Satisfaction'] = raw_data_two['Satisfaction'].map(satisfaction_mapping)
```

```python
gender_mapping = {'Male':0,'Female':1}
```

```python
raw_data_two['Gender'] = raw_data_two['Gender'].map(gender_mapping)
```

```python
customer_type_mapping = {'First-time':0,'Returning':1}
```

```python
 raw_data_two['Customer Type'] = raw_data_two['Customer Type'].map(customer_type_mapping)
```

```python
type_travel_mapping = {'Business':0,'Personal':1}
```

```python
raw_data_two['Type of Travel'] = raw_data_two['Type of Travel'].map(type_travel_mapping)
```

# <h1> REMOVING ROWS WITH SATISFACTION LEVEL = 0 ('not applicable')  <h1
                                                                       >

```python
columns_to_check = ['Ease of Online Booking',
       'Check-in Service', 'Online Boarding', 'Gate Location',
       'On-board Service', 'Seat Comfort', 'Leg Room Service',
       'Cleanliness', 'Food and Drink', 'In-flight Service',
       'In-flight Wifi Service', 'In-flight Entertainment',
       'Baggage Handling']
```

```python
for column in columns_to_check:
    zero_count = (raw_data_cmod[column] == 0).sum()
    print(f" {column}: {zero_count}")
```

```python
def remove_rows_with_zero_in_columns(df, columns_to_check):
  
    mask = df[columns_to_check].ne(0).all(axis=1)
    
    df_filtered = df[mask]
    
    return df_filtered
```

```python
raw_data_cmod = remove_rows_with_zero_in_columns(raw_data_cmod, columns_to_check)
```

```python
print(f"Rozmiar po: {raw_data_cmod.shape[0]}")
```

```python
for column in columns_to_check:
    zero_count = (raw_data_cmod[column] == 0).sum()
    print(f" {column}: {zero_count}")
```
